```python
import pandas as pd
import numpy as np
```

```python
### 9 AM QC AUDIT REPORT
qc_9=pd.read_excel('TN General Survey QC Report 9.xlsx')
qc_9=qc_9[['agent_id','qc_id']]
mm_9=qc_9.groupby('qc_id')['agent_id'].count()
mm_9=mm_9.to_frame()
mm_9.columns=['09:00']
sum_9=mm_9['09:00'].sum()
mm_9.loc["Grand Total"]=[sum_9]
mm_9["Total Calls Audited"]=mm_9['09:00']
mm_9=mm_9[['Total Calls Audited','09:00']]
mm_9.to_excel('9AM_QC_AUDIT_REPORT.xlsx')
```

```python
### 10 AM QC AUDIT REPORT
qc_10=pd.read_excel('TN General Survey QC Report 10.xlsx')
qc_10=qc_10[['agent_id','qc_id']]
grp_10=qc_10.groupby('qc_id')['agent_id'].count()
grp_10=grp_10.to_frame()
grp_10.columns=['10:00']
mm_10=pd.concat([mm_9,grp_10],axis=1)
mm_10['09:00']=mm_10['09:00'].replace(np.nan,0)
mm_10['10:00']=mm_10['10:00'].replace(np.nan,0)
mm_10['10:00']=mm_10['10:00']-mm_10['09:00']
mm_10["Total Calls Audited"]=mm_10['09:00']+mm_10['10:00']
tot_10=mm_10["Total Calls Audited"].sum()
sum_10=qc_10.groupby('qc_id')['agent_id'].count().sum()-sum_9
mm_10.loc["Grand Total"]=[tot_10,sum_9,sum_10]
data_10={"Total Calls Audited":mm_10["Total Calls Audited"],"09:00":mm_10['09:00'],"10:00":mm_10['10:00']}
data_10=pd.DataFrame(data_10)
data_10.to_excel('10AM_QC_AUDIT_REPORT.xlsx')
```
